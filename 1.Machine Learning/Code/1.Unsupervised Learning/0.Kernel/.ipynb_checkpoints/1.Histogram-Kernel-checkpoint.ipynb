# Histogram's kernel
This notebook will demonstrate how to draw a kernel density estimation

```python
# Import the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
```

```python
# We will need an additional function
# import the Kernel Density Estimator
from sklearn.neighbors import KernelDensity
```

Reading the data

```python
# Once again we read the CoCo data from our computer
data = pd.read_csv('../../../Data/CoCo.csv')
CoCo =  data['yield'].values
# CoCo is a list where each element is a yield value, To be consumed by the 
# kernel density estimator we need to reshape it to be a *list of lists* where each element
# is a value
# When using reshape, a "-1" means "unspecified/infer", i.e. the second dimention shall be one and the first dimention
# is chosen accordingly to fit the data
CoCo = CoCo.reshape(-1,1)
```

When calculating the Kernel Density, there are several kernels that we can use. For a detailed explanation see:

https://scikit-learn.org/stable/modules/density.html

#### TopHat Kernel

```python
# We can choos any kernel from [‘gaussian’|’tophat’|’epanechnikov’|’exponential’|’linear’|’cosine’] 
# Default is ‘gaussian’.
kde = KernelDensity(kernel='tophat')
kde.fit(CoCo)
```

```python
fig = plt.figure(figsize=(8,8))
# Create an array from 4 to 12 in intervals of .1 and reshape it
x = np.arange(4,12,0.01)
x = x.reshape(-1,1)
# Sample our kernel estimation on the range dictated by x
log_dens = kde.score_samples(x)
# reshape the result for plotting
log_dens = log_dens.reshape(-1,1)
# Since pur result is in a log scale we exponentieate it an then plot it
plt.plot(x, np.exp(log_dens),linewidth=2)
plt.xlabel('CoCo Yield');
plt.title('Top Hat Kernel');
```

### Gaussian Kernel with $\lambda=1$

```python
np.newaxis
```

```python
fig = plt.figure(figsize=(8,8))
# Define a 'gaussian' kernel with a bandwidth of 1 
kde = KernelDensity(kernel='gaussian',bandwidth=1)
# fit the kernel
kde.fit(CoCo)
# Sample the kernel over a range x
x = np.arange(4,12,0.01)
x = x.reshape(len(x),1)
log_dens = kde.score_samples(x)
# Reshape the result this time with a different trick using list indexing and the method np.newaxis
# the result is the same as x.reshape(-1,1)
log_dens = log_dens[:,np.newaxis]
plt.plot(x, np.exp(log_dens),linewidth=2);
plt.xlabel('CoCo Yield');
plt.title('Gaussian Kernel $\lambda =1$');
```

### Create kernel density for different bandwidths :
$[1.0,2.0,3.0]$

```python
fig = plt.figure(figsize=(8,8))
bw = np.array([1.0,2,3])
i=0
color = ['r','g','b']
# Iterate on each of the the band widths and plot a kernel with each on a different color
for b in bw:
    kde = KernelDensity(kernel='gaussian',bandwidth=b)
    kde.fit(CoCo)
    plt.plot(CoCo,np.exp(kde.score_samples(CoCo)),color[i],label='$\lambda=$'+str(b))
    i+=1
plt.legend();
    
```

### Finding the probability that yield between 7 and 8.<br>
The kernel can be used as random number generator.

```python
bw = np.array([1,2,3])
# Iterate on each badwidth
for b in bw:
    kde = KernelDensity(kernel='gaussian',bandwidth=b)
    kde.fit(CoCo)
    # take 1k samples from the current kernel and bandwidth
    v = kde.sample(n_samples=1000)
    print('Bandwidth' + str(b))
    # Calculate the probability as the frequenxy of the sampled values in the 
    # desired bin
    print('Prob=' + str(sum((v<8)&(v>7))/len(v)))
```

<h3>Finding the optimal bandwidth</h3>

<p> We will demonstrate a technique called grid search to find the badwidth that fits the best </p>

```python
# import the grid searcher from the module model_selection from sklearn
from sklearn.model_selection import GridSearchCV
```

Gridsearch will search over a range of different values and return the one that works the best

```python
# Define the search space
param = {'bandwidth': np.arange(0.01,3,0.005)}
# Initialize the grid search to look for the best parameter in "param" for KernelDensity()
grid = GridSearchCV(KernelDensity(),param)
# Fit the gridsearch to the CoCo data
grid.fit(CoCo)
```

```python
# Our variable grit is now fit to the data. we will retrieve the parameter that worked the best:
grid.best_estimator_.bandwidth
```
