# Supervised Learning: Principal Component Analysis

## Case Study: Studying the movement of a yield curve

The history of the EUR-swap curve across 8 different tenors is provided in the file _Yield.csv_ .
What are the principal components best describing the daily changes in the EUR yield curve?

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.decomposition import PCA
%matplotlib inline
```

### Read the data from csv file

```python
df = pd.read_csv('../../../Data/Yield.csv',parse_dates=True,index_col='T')
# Change the index's name to T
df.index.name= 'T'
# Print the first 5 rows
df.head()
```

## Plot the tenor values over time
### Time series with 8 tenors

```python
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

fig = plt.figure(figsize=(8,8))
for c in df.columns:
    plt.plot(df.index, df[c].values, label=c)
plt.legend()
plt.xlabel('T',fontsize=14)
plt.ylabel('R (%)',fontsize=14)
plt.title('EUR Swap Rates',fontsize=20);
```

### Run PCA with all components

```python
d = df.shape[1]
pca = PCA(n_components=d)
pca.fit(df.values)
```

## Variance explained per eigenvalue

### Plot the Eigenvalues

```python
fig = plt.figure(figsize=(8,8))
x = np.arange(1, d+1)
y = pca.singular_values_**2
plt.bar(x, y, linewidth=2)
plt.xlabel('Component',fontsize=14)
plt.ylabel('Eigenvalue ($\lambda$)',fontsize=14);
```

### Plot explained variance (%) per eigenvalue

```python
fig = plt.figure(figsize=(8,8))
x = np.arange(1, d+1)
y = 100*pca.explained_variance_ratio_
plt.bar(x, y, linewidth=2)
plt.xlabel('Component',fontsize=14)
plt.ylabel('Variance explained (%)',fontsize=14);
```

## Total variance explained per eigenvalue

```python
fig = plt.figure(figsize=(8,8))
x = np.arange(1, d+1)
y = np.cumsum(100*pca.explained_variance_ratio_)
plt.plot(x, y, linewidth=2)
plt.scatter(x, y)
plt.xlabel('Component',fontsize=14)
plt.ylabel('Total variance explained (%)',fontsize=14);
```

## First eigenvector: parallel shift of yield curve

```python
fig = plt.figure(figsize=(8,8))
x = df.columns
y = pca.components_[0, :]
plt.bar(x, y, linewidth=2)
plt.xlabel('Tenor',fontsize=14)
plt.ylabel('Eigenvector',fontsize=14);
```

## Second eigenvector: steepening/flattening of the curve

```python
fig = plt.figure(figsize=(8,8))
x = df.columns
y = pca.components_[1, :]
plt.bar(x, y, linewidth=2)
plt.xlabel('Tenor',fontsize=14)
plt.ylabel('Eigenvector',fontsize=14);
```

## Dimension reduction: using 2 components $(Z_{p})$

```python
# Defiinf 2 components for the analysis
n_components = 2

# PCA implementation
pca = PCA(n_components=n_components)
x_reduced = pca.fit_transform(df.values)
df_reduced = pd.DataFrame(x_reduced, index=df.index)
# Display the first 5 rows of the reduced data frame
df_reduced.head()
```

```python
fig = plt.figure(figsize=(8,8))
for c in df_reduced.columns:
    plt.plot(df_reduced.index, df_reduced[c].values, label=c)
plt.legend()
plt.xlabel('T',fontsize=14)
plt.title('Components',fontsize=20);
```

# Smoothed swap rates: using 2 components $(X_{p})$

```python
x_smooth = x_reduced.dot(pca.components_) + pca.mean_
df_smooth = pd.DataFrame(x_smooth, index=df.index, columns=df.columns)
df_smooth.head()
```

```python
fig = plt.figure(figsize=(8,8))
for c in df_smooth.columns:
    plt.plot(df_smooth.index, df_smooth[c].values, label=c)
plt.legend()
plt.xlabel('T',fontsize=14)
plt.ylabel('R (%)',fontsize=14)
plt.title('EUR Swap Rates (Smoothed)',fontsize=20);
```
